are totally different from those of a resistor.

change in the voltage or current is made in the network.

### Capacitor Voltage

The voltage across a capacitor does not change instantaneously This
voltage begins as zero and gradually climbs to full voltage. Full
voltage is the source voltage. This change may range from nanoseconds to
milliseconds depending on the resistance and capacitance.

### Capacitor Current

During charging, electrons move from one plate to another. This current
lasts only until the capacitor is charged. The current goes from a large
initial spike to zero. No current passes through the dielectric.

CAPACITORS IN SERIES AND IN PARALLEL

Capacitors, like resistors, can be placed in series and in parallel.
Increasing levels of capacitance can be obtained by placing capacitors
in parallel, while decreasing levels can be obtained by placing
capacitors in series.
