## Exponential Functions

Let $b$ be a constant real number such that $b>0$ and $b\neq1$

Then for any real number $x$ ,a function of the form

$$f(x)=b^{x}$$

is called an exponential function of base $b$

$a^{-n}=\dfrac{1}{a^{n}}$

$a^{n}=\dfrac{1}{a^{-n}}$

$\left(\dfrac{a}{b}\right)^{-n}=\left(\dfrac{b}{a}\right)^{n}$

1.  Isolate the exponential expression on one side of the equation.

2. Take a logarithm of the same base on both sides of the equation.

3. Use the power property of logarithms to “bring down” the exponent.

4. Solve the resulting equation.
