### The Difference Quotient

The average rate of change between $P$ and $Q$ is the slope of the
secant line and is given by

$$m=\frac{\Delta y}{\Delta x}=\frac{f(x+h)-f(x)}{(x+h)-x}=\frac{f(x+h)-f(x)}{h}$$

The expression $\frac{f(x+h)-f(x)}{h}$ is called the difference
quotient. In Calculus, itis alled the “secant line”.

You caninterpret the

difference quotient as the average rate of change of your function
between two points.



As a picture, it’s the slope of the line connecting two points of your
function.
