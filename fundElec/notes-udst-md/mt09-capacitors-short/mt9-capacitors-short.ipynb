### Capacitor Voltage

- The voltage across a capacitor does not change instantaneously.
- This voltage begins as zero and gradually climbs to full voltage.
- Full voltage is the source voltage.
- This change may range from nanoseconds to milliseconds depending on the resistance and capacitance.

### Capacitor Current

- During charging, electrons move from one plate to another.
- This current lasts only until the capacitor is charged.
- The current goes from a large initial spike to zero.
- No current passes through the dielectric.

# Capacitors in Parallel

For capacitors in parallel:

The total charge $Q_T$ is the sum of the charges on each individual capacitor:
$$Q_T = Q_1 + Q_2 + Q_3$$

The voltage $E$ across the parallel combination is the same as the voltage across each individual capacitor:
$$E = V_1 = V_2 = V_3$$

Using $Q = CV$, we can write:
$$C_T E = C_1 V_1 = C_2 V_2 = C_3 V_3$$

Since $E = V_1 = V_2 = V_3$, we can divide through by the voltage:
$$C_T = C_1 + C_2 + C_3$$

Thus, the total capacitance of capacitors in parallel is given by:
$$C_T = C_1 + C_2 + C_3 + \dots + C_n$$
